# Импорт библиотек

```python
import uirs_env
import stable_baselines3 as sb
from stable_baselines3.common.env_checker import check_env
from stable_baselines3.common.callbacks import CheckpointCallback
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common import results_plotter
import numpy as np
import matplotlib.pyplot as plt
from stable_baselines3.common.results_plotter import load_results, ts2xy
import random
import os
base_env = uirs_env.RobotinoWorldEnv()

checkpoint_callback = CheckpointCallback(
  save_freq=10000,
  save_path="./logs/",
  name_prefix="rl_model_ppo",
  save_replay_buffer=True,
  save_vecnormalize=True,
)

log_dir = "./logs/"
os.makedirs(log_dir, exist_ok=True)

# print(base_env.field.pth[0],base_env.field.pth[1],base_env.field.pth[2])
```

# ВАЖНО! Перед началом обучения стоит задать траекторию, иначе возможны баги

Установка траектории

```python
base_env.set_traj([(400,300),(300,280),(500,240), (500, 400)])
```

Создание необученной модели

```python
model = sb.A2C('MlpPolicy', base_env, verbose=1)
```

Обучение модели и её сохранение

```python
model.learn(4e4, callback=checkpoint_callback, progress_bar=False)
model.save('a2c_new_model')
```

Загрузка обученной модели

```python
model = sb.A2C.load('a2c_new_model',base_env)
```

Запуск env без управления нейросетью, с константной уставкой.

```python
state, info = base_env.reset()
terminated = False
truncated = False
total = 0
timesteps = 0
#print(env.field.wheel1.posX, env.field.wheel1.posY)
while terminated == False and truncated == False:
   #action, _state = model.predict(state, deterministic = True)
   state, reward, terminated, truncated, info = base_env.step([0,1,0])
   base_env.render()
   #print(round(base_env.field.move_angle,4), '\t\t\t\t', round(base_env.field.obj_angle,4))
   print(base_env.field.Vy, "\t\t\t\t") #base_env.field.dif_angle
   #print(reward)
   total += reward
print("total reward ", total)
base_env.close()
```

Запуск обученной модели

```python
state, info = base_env.reset()
terminated = False
truncated = False
total=0
timesteps = 0
#print(env.field.wheel1.posX, env.field.wheel1.posY)and truncated == False
while terminated == False and truncated == False:
   timesteps += 1
   action, _state = model.predict(state)
   state, reward, terminated, truncated, info = base_env.step(action)
   total += reward
   base_env.render()
   #print(round(base_env.field.prevrange,2),"\t\t\t\t\t\t", round(base_env.field.prange,2))
   print(reward)
   
print("total reward ", total)
print("total timesteps ", timesteps)
base_env.close()
```
